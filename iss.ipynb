This notebook collects ISS location data (lat/lon/timestamp) using the Open Notify API,
stores at least 100 points, and visualizes the path with matplotlib.


```python
import requests
import time
import matplotlib.pyplot as plt
```

```python
latitudes, longitudes, timestamps = [], [], []

for i in range(100):
    r = requests.get("http://api.open-notify.org/iss-now.json").json()
    latitudes.append(float(r["iss_position"]["latitude"]))
    longitudes.append(float(r["iss_position"]["longitude"]))
    timestamps.append(r["timestamp"])
    print(f"Point {i+1}: lat={latitudes[-1]}, lon={longitudes[-1]}, time={timestamps[-1]}")
    time.sleep(10)
```

```python
plt.figure(figsize=(10,6))
plt.plot(longitudes, latitudes, marker="o", markersize=3, linestyle="--", color="blue")
plt.title("ISS Path Across the Earth")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
